```python
from sklearn.datasets import load_iris

X,y = load_iris(return_X_y=True)
```

```python
from sklearn.preprocessing import Normalizer

normalize = Normalizer()

X = normalize.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

train_x,test_x,train_y,test_y = train_test_split(X,y,test_size=0.2,stratify=y,random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5,weights="uniform")

model = knn.fit(train_x,train_y)
```

```python
pred_y = model.predict(test_x)
```

```python
import matplotlib.pyplot as plt

plt.subplots(figsize=(10,4))

plt.subplot(1,2,1)
plt.title("Predicted Label")
plt.xlabel("No. of samples")
plt.ylabel("Classes")
plt.scatter(x=range(0,30),y=pred_y,marker="+",c="black")

plt.subplot(1,2,2)
plt.title("Actual Label")
plt.xlabel("No. of samples")
plt.ylabel("Classes")
plt.scatter(x=range(0,30),y=test_y,marker="*",c="black")
```

```python
from sklearn import metrics

print(f"Accuracy Score: {metrics.accuracy_score(test_y,pred_y)*100}")
print(f"Precision Score: {metrics.precision_score(test_y,pred_y,average='weighted')*100}")
print(f"Recall Score: {metrics.recall_score(test_y,pred_y,average='weighted')*100}")
```

```python
cm = metrics.confusion_matrix(test_y,pred_y)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=['setosa','virginica','versicolor'])
cm_display.plot(cmap='plasma')
```
